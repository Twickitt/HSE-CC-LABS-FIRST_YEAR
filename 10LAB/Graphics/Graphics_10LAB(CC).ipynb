```python
import matplotlib.pyplot as plt 
import pandas as pd


file = pd.read_csv("Data.csv", sep = ',', header = None)

file.columns = ['Size', 'Hash_time1', 'Collisions1', 'Hash_time2', 'Collisions2', 'Hash_time3', 'Collisions3', 'Hash_time4', 'Collisions4']


plt.figure(figsize = (12, 6))

for i in range(1,5):
    plt.plot(file["Size"], file[f"Hash_time{i}"], label = f"Hash_Time{i}", marker = 'o', linewidth = 2, markersize=5)

plt.yscale('log')
plt.xlabel('Size(log scale)')
plt.ylabel('Time')
plt.title('Searching time of each Hash function depending on size')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('Graphic_Of_Searching_Time.png')
plt.show()


plt.figure(figsize = (12, 6))

for i in range(1,5):
    plt.plot(file["Size"], file[f"Collisions{i}"], label = f"Collisions{i}", marker = 'o', linewidth = 2, markersize=5)

plt.yscale('log')
plt.xlabel('Size(log scale)')
plt.ylabel('Collisions')
plt.title('Collisions of each function depending on size')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('Graphic_Of_Collisions.png')
plt.show()
```
